```python
df['Category'].value_counts().plot(kind='pie', autopct='%1.1f%%')
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df['MRP'] = pd.to_numeric(df['MRP'].str.replace('[₹,]', '', regex=True), errors='coerce')
df['SellPrice'] = pd.to_numeric(df['SellPrice'].str.replace('[₹,]', '', regex=True), errors='coerce')
df['Discount'] = pd.to_numeric(df['Discount'].str.replace('[%]', '', regex=True), errors='coerce')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['SellPrice'])
```

```python
df['MRP'] = df['MRP'].fillna(df['SellPrice'])
```

```python
df['Discount'] = df['Discount'].fillna(
    ((df['MRP'] - df['SellPrice']) / df['MRP'] * 100).round(2)
)
```

```python
df['Sizes'] = df['Sizes'].replace('Nan', pd.NA)
df['Sizes'] = df['Sizes'].fillna('Not Specified')
```

```python
print(df.isnull().sum())
```

```python
print(f"\nRemaining rows: {len(df)}")
```

```python
df.groupby("BrandName")["SellPrice"].mean()
```

```python
total_revenue = df['SellPrice'].sum()
```

```python
total_orders = len(df)
```

```python
aov = df['SellPrice'].mean()
```

```python
print(f"Total Revenue  : ₹{total_revenue:,.2f}")
```

```python
print(f"🛒 Total Orders   : {total_orders}")
```

```python
print(f"📊 AOV            : ₹{aov:,.2f}")
```

Add Filters — Category & Brand

```python
category_revenue = df.groupby('Category')['SellPrice'].agg(
    Total_Revenue='sum',
    Total_Orders='count',
    AOV='mean'
).round(2).sort_values('Total_Revenue', ascending=False)
```

```python
print(category_revenue)
```

```python
brand_revenue = df.groupby('BrandName')['SellPrice'].agg(
    Total_Revenue='sum',
    Total_Orders='count',
    AOV='mean'
).round(2).sort_values('Total_Revenue', ascending=False).head(10)
```

```python
print(brand_revenue)
```

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(16, 10))
fig.suptitle('Fashion Sales Dashboard', fontsize=20, fontweight='bold', color='purple')

gs = gridspec.GridSpec(2, 2, figure=fig)

# --- KPI Cards ---
ax0 = fig.add_subplot(gs[0, :])
ax0.axis('off')
kpis = [
    f"💰 Total Revenue\n₹{total_revenue:,.0f}",
    f"🛒 Total Orders\n{total_orders:,}",
    f"📊 AOV\n₹{aov:,.2f}"
]
for i, kpi in enumerate(kpis):
    ax0.text(0.15 + i*0.3, 0.5, kpi, ha='center', va='center',
             fontsize=14, fontweight='bold',
             bbox=dict(boxstyle='round,pad=1', facecolor='#f3e5f5', edgecolor='purple'))

# --- Revenue by Category ---
ax1 = fig.add_subplot(gs[1, 0])
category_revenue['Total_Revenue'].head(8).plot(
    kind='barh', ax=ax1, color='mediumpurple')
ax1.set_title('Revenue by Category', fontweight='bold')
ax1.set_xlabel('Revenue (₹)')

# --- Top 10 Brands ---
ax2 = fig.add_subplot(gs[1, 1])
brand_revenue['Total_Revenue'].plot(
    kind='bar', ax=ax2, color='plum')
ax2.set_title('Top 10 Brands by Revenue', fontweight='bold')
ax2.set_ylabel('Revenue (₹)')
ax2.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig('fashion_dashboard.png', dpi=150, bbox_inches='tight')
plt.show()
```
